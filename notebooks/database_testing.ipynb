## Create database connection

 - **Database Connection:**  
   - **mysql:** Database
   - **pymysql:** Database connector
   - **root:** Username
   - **toor:** Password
   - **@127.0.0.1:3306** - IP
   - **dota2learning:** Database name

````python
from sqlalchemy.engine import create_engine
````

````python
engine = create_engine('mysql+pymysql://root:toor@127.0.0.1:3306/dota2learning')
````

## 01 - Testing "SELECT * FROM hero"

````python
select_query = """
SELECT * FROM hero
"""
````

````python
import pandas as pd
heroes = pd.read_sql_query(select_query, engine).set_index('id')
````

**NOTE:**  
**set_index('id')** define **id** as index column.
